```python
import pandas as pd

# Veri setini yÃ¼kle
df = pd.read_csv("psychology_10k_cleaned.csv").head(1000)

# Veri setini incele
print(df.head())
print(f"Veri seti boyutu: {df.shape}")
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

model_name = "distilgpt2"  
model = AutoModelForCausalLM.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
from datasets import Dataset

# Pandas DataFrame'i Hugging Face Dataset formatÄ±na dÃ¶nÃ¼ÅŸtÃ¼r
dataset = Dataset.from_pandas(df)
# Tokenizer'Ä±n pad_token'Ä±nÄ± ayarla
tokenizer.pad_token = tokenizer.eos_token


def tokenize_function(examples):
    inputs = [f"User: {text} {tokenizer.eos_token}" for text in examples["input"]]
    outputs = [f"Bot: {text} {tokenizer.eos_token}" for text in examples["output"]]

    model_inputs = tokenizer(inputs, max_length=32, truncation=True, padding="max_length")  # max_length=32
    labels = tokenizer(outputs, max_length=32, truncation=True, padding="max_length")["input_ids"]

    model_inputs["labels"] = labels
    return model_inputs


# Tokenize edilmiÅŸ veri setini oluÅŸtur
tokenized_dataset = dataset.map(tokenize_function, batched=True)
```

```python
from transformers import Trainer, TrainingArguments
import numpy as np
import evaluate

# Accuracy metriÄŸini tanÄ±mla
accuracy = evaluate.load("accuracy")


def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return accuracy.compute(predictions=predictions, references=labels)

# EÄŸitim parametreleri
training_args = TrainingArguments(
    output_dir="./results",                   # GeÃ§ici Ã§Ä±ktÄ± dizini
    evaluation_strategy="steps",             # Belirli adÄ±mlarda doÄŸrulama
    eval_steps=500,                          # 500 adÄ±mda bir doÄŸrulama
    logging_steps=500,                       # 500 adÄ±mda bir loss/accuracy loglama
    save_strategy="no",                      # Checkpoint kaydetme devre dÄ±ÅŸÄ±
    per_device_train_batch_size=2,           # GPU baÅŸÄ±na batch boyutu
    per_device_eval_batch_size=2,            # DoÄŸrulama iÃ§in batch boyutu
    learning_rate=3e-5,                      # Ã–ÄŸrenme hÄ±zÄ±
    num_train_epochs=3,                      # EÄŸitim epok sayÄ±sÄ±
    warmup_steps=500,                        # Ã–ÄŸrenme hÄ±zÄ±nÄ± artÄ±rma adÄ±mlarÄ±
    weight_decay=0.01,                       # AÄŸÄ±rlÄ±k azalmasÄ± (regularization)
    logging_dir='./logs',                    # Loglar iÃ§in dizin
    report_to="none",                        # Sadece terminal Ã§Ä±ktÄ±sÄ±
    fp16=True,  # Mixed precision
)

# Trainer ile eÄŸitim
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,         # EÄŸitim veri seti
    eval_dataset=tokenized_dataset,          # DoÄŸrulama veri seti (eÄŸer ayrÄ± varsa ekle)
    tokenizer=tokenizer,
    compute_metrics=compute_metrics,         # DoÄŸrulama iÃ§in metrikler
)

# EÄŸitimi baÅŸlat
trainer.train()
```
