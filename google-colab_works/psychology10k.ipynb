```python
import pandas as pd
from datasets import load_dataset
# Veri setini indir
dataset = load_dataset("samhog/psychology-10k")

# Veri setinin ilk 5 satırını incele
print(dataset['train'][0])
```

```python
import pandas as pd

# Tüm veri setini DataFrame'e dönüştürme
df = pd.DataFrame(dataset['train'])
```

```python
# 'instruction' sütununu kaldır
df = df.drop(columns=['instruction'])

# Veri setini incele
print(df.head())
```

```python
# Temizleme işlemleri
import re

def clean_text(text):
    text = re.sub(r'[^\w\s]', '', text)  # Noktalama işaretlerini kaldırma
    text = text.lower()  # Küçük harfe dönüştürme
    return text

df['input'] = df['input'].apply(clean_text)
df['output'] = df['output'].apply(clean_text)

# İşlenmiş veri setini CSV formatında kaydet
df.to_csv("psychology_10k_cleaned.csv", index=False)

print("Veri seti işlenmiş ve 'psychology_10k_cleaned.csv' olarak kaydedilmiştir.")
```

```python
import pandas as pd

# Veri setini yükle
df = pd.read_csv("psychology_10k_cleaned.csv").head(1000)

# Veri setini incele
print(df.head())
print(f"Veri seti boyutu: {df.shape}")
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

model_name = "distilgpt2"  
model = AutoModelForCausalLM.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
from datasets import Dataset

# Pandas DataFrame'i Hugging Face Dataset formatına dönüştür
dataset = Dataset.from_pandas(df)
# Tokenizer'ın pad_token'ını ayarla
tokenizer.pad_token = tokenizer.eos_token


def tokenize_function(examples):
    inputs = [f"User: {text} {tokenizer.eos_token}" for text in examples["input"]]
    outputs = [f"Bot: {text} {tokenizer.eos_token}" for text in examples["output"]]

    model_inputs = tokenizer(inputs, max_length=32, truncation=True, padding="max_length")  # max_length=32
    labels = tokenizer(outputs, max_length=32, truncation=True, padding="max_length")["input_ids"]

    model_inputs["labels"] = labels
    return model_inputs


# Tokenize edilmiş veri setini oluştur
tokenized_dataset = dataset.map(tokenize_function, batched=True)
```

```python
from transformers import Trainer, TrainingArguments
import numpy as np
import evaluate

# Accuracy metriğini tanımla
accuracy = evaluate.load("accuracy")


def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return accuracy.compute(predictions=predictions, references=labels)

# Eğitim parametreleri
training_args = TrainingArguments(
    output_dir="./results",                   # Geçici çıktı dizini
    evaluation_strategy="steps",             # Belirli adımlarda doğrulama
    eval_steps=500,                          # 500 adımda bir doğrulama
    logging_steps=500,                       # 500 adımda bir loss/accuracy loglama
    save_strategy="no",                      # Checkpoint kaydetme devre dışı
    per_device_train_batch_size=2,           # GPU başına batch boyutu
    per_device_eval_batch_size=2,            # Doğrulama için batch boyutu
    learning_rate=3e-5,                      # Öğrenme hızı
    num_train_epochs=3,                      # Eğitim epok sayısı
    warmup_steps=500,                        # Öğrenme hızını artırma adımları
    weight_decay=0.01,                       # Ağırlık azalması (regularization)
    logging_dir='./logs',                    # Loglar için dizin
    report_to="none",                        # Sadece terminal çıktısı
    fp16=True,  # Mixed precision
)

# Trainer ile eğitim
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,         # Eğitim veri seti
    eval_dataset=tokenized_dataset,          # Doğrulama veri seti (eğer ayrı varsa ekle)
    tokenizer=tokenizer,
    compute_metrics=compute_metrics,         # Doğrulama için metrikler
)

# Eğitimi başlat
trainer.train()
```
