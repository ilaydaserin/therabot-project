```python
import json
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from transformers import DistilBertTokenizer, DistilBertForSequenceClassification, Trainer, TrainingArguments
from torch.utils.data import Dataset
import torch
```

```python
# JSON dosyasını yükle
with open('intents.json', 'r') as f:
    intents = json.load(f)
```

```python
# Metinler ve etiketleri ayrıştır
texts = []
labels = []

for intent in intents:
    examples = intent["examples"]
    intent_label = intent["intent"]
    texts.extend(examples)
    labels.extend([intent_label] * len(examples))
```

```python
texts_augmented = []
labels_augmented = []

for text, label in zip(texts, labels):
    # Her örneği en az 5 kez çoğalt
    for _ in range(5):
        texts_augmented.append(text)
        labels_augmented.append(label)

# Güncellenmiş veriyi kullanarak devam 
texts = texts_augmented
labels = labels_augmented

# Sınıf dağılımını kontrol edin
from collections import Counter
class_counts = Counter(labels)
print("Güncellenmiş sınıf dağılımı:", class_counts)
print(f"Güncellenmiş veri sayısı: {len(texts)}")
```
